```python
# Comparer les colonnes'target' avec 'sex' pour connaitre comment la variable independent 'genre' interagir avec la variable dependant 'target'
pd.crosstab(Data.target, Data.genre)
#on a 96 femmes et 207 hommes dans cette dataset
#basé seulement sur cette comparation nous pouvons dire que les femmes ont un pourcentage 75% d'avoir une maladie cardique, cependant les hommes ont un pourcentage de 50%
```

```python
# nous allong illustrer cette comparaison avec une graph bar
pd.crosstab(Data.target, Data.genre).plot(kind="bar", figsize=(10,6), color=["m", "y"])

plt.title(" Fréquences de maladies cardiaques pour les deux genres")
plt.xlabel("0 = Maladie négative, 1 = Maladie positive")
plt.ylabel("Fréquence")
plt.legend(["Femme", "Homme"])
```

```python
#mtn on combine deux variables independantes comme par exemple age and fcm (fréquence cardiaque maximale) et les comparer avec Genre
# creation de la figure

plt.figure(figsize=(10,6))

# on commence par Homme
plt.scatter(Data.age[Data.genre==1], 
            Data.fcm[Data.genre==1], 
            c="y") # pour qu il s'agit de scatter plot

# mtn pour Femme, on utilise plt pour que les deux graphs soient dans la mm figure
plt.scatter(Data.age[Data.genre==0], 
            Data.fcm[Data.genre==0], 
            c="m") 


plt.title("Le Genre en fonction de l'Age et la Fréquence cardiaque maximal")
plt.xlabel("Age")
plt.legend(["Homme", "Femme"])
plt.ylabel("Fréquence cardiaque maximale")

#Nous ne pouvons tirer aucune informations àpartir de cette graph
```

```python

#mtn on combine deux variables independantes comme par exemple age and fcm (fréquence cardiaque maximale) et les comparer avec notre variable dependante
#on plusieurs valeurs des variables age et fcm donc nous allong faire une figure avec scatter plot
# creation de la figure

plt.figure(figsize=(10,6))

# on commence par ceux qui sont positive
plt.scatter(Data.age[Data.target==1], 
            Data.fcm[Data.target==1], 
            c="k", marker = 'x') # pour qu il s'agit de scatter plot

# mtn pour les négatives, on utilise plt pour que les deux graphs soient dans la mm figure
plt.scatter(Data.age[Data.target==0], 
            Data.fcm[Data.target==0], 
            c="g") 


plt.title("Maladie cardiaque en fonction de l'Age et la Fréquence cardiaque maximal")
plt.xlabel("Age")
plt.legend(["Maladie", "No maladie"])
plt.ylabel("Fréquence cardiaque maximale")

# on constate que plus l'age augmente plus la Fréquence cardiaque maximale diminue 
```

```python
#teston encore une autre variable independante mtn on va utiliser td (type de la douleur thoracique)
#on fait le mm proces comme on a fait avec la variable independante age
pd.crosstab(Data.td, Data.target)
```

```python
#mtn on visualise les Fréquences de la Maladie cardiaque Par type de la douleur thoracique
#on visualise avec bar car on a des variables categorialle facile a visualiser avec bar
pd.crosstab(Data.td, Data.target).plot(kind="bar", 
                                   figsize=(10,6), 
                                   color=["g", "k"])


plt.title("Fréquence de la Maladie cardiaque Par type de la douleur thoracique")
plt.xlabel("type de la douleur thoracique")
plt.ylabel("Fréquence")
plt.legend(["No Maladie", "Maladie"])
#mm si ma valeur 1 : n'a aucun relation avec le coeur , on constate qu il a une influence sur lui
#la valeur 1 est : Angor atypique: douleur thoracique non liée au cœur
```

```python
#mtn on combine deux variables independantes comme par exemple age and chol (fréquence cardiaque maximale) et les comparer avec notre variable dependante target
# creation de la figure

plt.figure(figsize=(10,6))

# on commence par les cas positives
plt.scatter(Data.age[Data.target==1], 
            Data.fcm[Data.target==1], 
            c="k", marker = 'x') # pour qu il s'agit de scatter plot


# mtn pour les cas négatives, on utilise plt pour que les deux graphs soient dans la mm figure
plt.scatter(Data.age[Data.target==0], 
            Data.fcm[Data.target==0], 
            c="g") 


plt.title("Maladie Cardiovasculaire en fonction des Niveaux du Cholesterole")
plt.xlabel("Age")
plt.legend(["Maladie", "Non Maladie"])
plt.ylabel("Cholesterole")

#Nous ne pouvons tirer aucune informations àpartir de cette graph
```

```python
Data.fcm.plot(kind = 'line', color = 'g', label = 'Fréquence Card. max ',linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
Data.chol.plot(color = 'r',label = 'Cholésterol',linewidth = 1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
plt.ylabel('Valeur')
plt.title('Fréquence Cardiaque Maximale - Cholésterol')
plt.show()
```
